```python
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
import numpy as np
```

Read the data

```python
df = pd.read_csv("2015_trip_data.csv")
```

Change starttime, stoptime to datetime format

```python
df['starttime'] = pd.to_datetime(df['starttime'])
df['stoptime'] = pd.to_datetime(df['stoptime'])
```

Add column to df,'dayofweek.starttime','dayofweek.stoptime', 'startday', 'stopday'

```python
df['startday'] = (df['starttime']).dt.date
df['stopday'] = (df['stoptime']).dt.date
df['dayofweek.starttime'] = (df['starttime']).dt.weekday_name
df['dayofweek.stoptime'] = (df['stoptime']).dt.weekday_name
```

calculate from and to station counts by day of week.

```python
from_counts_by_dayofweek = df.groupby(['dayofweek.starttime','from_station_id']).size()
to_counts_by_dayofweek = df.groupby(['dayofweek.stoptime','to_station_id']).size()
```

```python
numdays_start = df.groupby(['dayofweek.starttime'])['startday'].nunique()
numdays_stop = df.groupby(['dayofweek.stoptime'])['stopday'].nunique()
```

calculate averages by day of week

```python
from_counts_by_dayofweek = from_counts_by_dayofweek/numdays_start
to_counts_by_dayofweek = to_counts_by_dayofweek/numdays_stop
```

Create a dataframe with average counts

```python
new_df = pd.DataFrame({'from':from_counts_by_dayofweek, 'to':to_counts_by_dayofweek})
```

```python
new_df.head()
```

Creating a new dataframe with the difference in from and To counts

```python
delta = (new_df['from'] - new_df['to']).unstack(level=0)
```

```python
delta.head()
```

Plotting the subplots using subplot functionality taught in class. ToDo: Not able to increase the X and Y axis size.

```python
plt.figure(figsize=(20, 40))
delta.plot(kind='bar', subplots=True)
```

Creating a function to create a bar plot for single column

```python
def plot_bar1(df, column, opts):
    """
    Does a bar plot for a single column.
    :param pd.DataFrame df:
    :param str column: name of the column to plot
    :param dict opts: key is plot attribute
    """
    n_groups = len(df.index)
    index = np.arange(n_groups)  # The "raw" x-axis of the bar plot
    rects1 = plt.bar(index, df[column])
    if 'xlabel' in opts:
      plt.xlabel(opts['xlabel'])
    if 'ylabel' in opts:
      plt.ylabel(opts['ylabel'])
    if 'xticks' in opts and opts['xticks']:
      plt.xticks(index, df.index)  # Convert "raw" x-axis into labels
      _, labels = plt.xticks()  # Get the new labels of the plot
      plt.setp(labels, rotation=90)  # Rotate labels to make them readable
    else:
      labels = ['' for x in df.index]
      plt.xticks(index, labels)   
    if 'ylim' in opts:
      plt.ylim(opts['ylim'])
    if 'title' in opts:
      plt.title(opts['title'])
```

Creating function to create bar plots for multiple columns

```python
def plot_barN(df, columns, opts):
    """
    Does a bar plot for a single column.
    :param pd.DataFrame df:
    :param list-of-str columns: names of the column to plot
    :param dict opts: key is plot attribute
    """
    num_columns = len(columns)
    local_opts = dict(opts)  # Make a deep copy of the object
    idx = 0
    for column in columns:
        idx += 1
        local_opts['xticks'] = False
        local_opts['xlabel'] = ''
        if idx == num_columns:
          local_opts['xticks'] = True
          local_opts['xlabel'] = opts['xlabel']
        plt.subplot(num_columns, 1, idx)
        plot_bar1(df, column, local_opts)
        plt.title(opts['title'][idx-1])
```

```python
fig = plt.figure(figsize=(20, 40))  # Controls global properties of the bar plot
opts = {'xlabel': 'Stations', 'ylabel': 'Delta Average Counts', 'ylim':[-10,10],'title': ['Sunday','Monday','Tuesday','Wednesday','Thursday','Friday','Saturday']}
plot_barN(delta,['Sunday','Monday','Tuesday','Wednesday','Thursday','Friday','Saturday'], opts)
```
